# 1. Introdução

A rotatividade de clientes refere-se ao fenômeno em que os clientes descontinuam seu relacionamento ou assinatura com uma empresa ou provedor de serviços. Representa a taxa na qual os clientes param de usar os produtos ou serviços de uma empresa em um período específico. O churn é uma métrica importante para as empresas, pois afeta diretamente a receita, o crescimento e a retenção de clientes. No contexto do conjunto de dados Churn, o rótulo churn indica se um cliente cancelou ou não. Um cliente cancelado é aquele que decidiu descontinuar sua assinatura ou uso dos serviços da empresa. Por outro lado, um cliente não cancelado é aquele que continua engajado e mantém seu relacionamento com a empresa.

# 2. Objetivo

Construir modelos preditivo para prever a rotatividade de clientes numa determinada empresa. Ou seja, com base nas características dos clientes prever se eles irãocancelar ou não o serviço.

# 3. Justificativa

Entender a rotatividade de clientes é crucial para as empresas identificarem padrões, fatores e indicadores que contribuem para a perda de clientes. Ao analisar o comportamento do churn e seus recursos associados, as empresas podem desenvolver estratégias para reter os clientes existentes, melhorar a satisfação do cliente e reduzir a rotatividade de clientes. As técnicas de modelagem preditiva também podem ser aplicadas para prever e lidar proativamente com a rotatividade potencial, permitindo que as empresas tomem medidas proativas para reter clientes em risco.

###### 8.2.1.2 Detectando outiliers com o algoritmo KNN

* Como estamos trabalhando com o KNN basicamente é feito um cálculo de distância entre os registros. Aqueles registros que tiver uma distância muito grande da maioria é considerado um outliers.

```python
# Obtendo as previsões
# Ele vai gerar a previsão para cada um dos registros indicando se é ou não é um outliers

previsoes = detector.labels_
previsoes
```

* Perceba que ele retorna um array com 0 e 1. O Zero indica que o registro não é um outliers e 1 indica que o registro é um outliers.

```python
# Contagem de quantos outliers foram detectados (ele vai contar quantos 0 e quantos 1)

np.unique(previsoes, return_counts=True)
```

```python
# Obtendo a confiança dessas previsões

confianca_previsoes = detector.decision_scores_
confianca_previsoes
```

```python
# Buscando os índices de registros que são considerados outliers

outliers = []
for i in range(len(previsoes)):
  #print(i)
  if previsoes[i] == 1:
    outliers.append(i)
```

```python
# Função para executar um coeficiente de correlação com variáveis categóricas

def cramer_v( x, y ):
    cm = pd.crosstab( x, y ).to_numpy()
    n = cm.sum()
    r, k = cm.shape

    chi2 = ss.chi2_contingency( cm )[0]
    chi2corr = max( 0, chi2 - (k-1)*(r-1)/(n-1) )

    #Corrigindo o viés V de Cramer
    kcorr = k - (k-1)**2/(n-1)
    rcorr = r - (r-1)**2/(n-1)

    return np.sqrt( (chi2corr/n) / ( min( kcorr-1, rcorr-1 ) ) )
```

```python
# Obtendo apenas dados categóricos
a = base_dados_clientes2.select_dtypes( include='object' )

# Calculando cramer V
a1 = cramer_v( a['Gênero'], a['Gênero'] )
a2 = cramer_v( a['Gênero'], a['Tipo_de_assinatura'] )
a3 = cramer_v( a['Gênero'], a['Duração_do_contrato'] )


a4 = cramer_v( a['Tipo_de_assinatura'], a['Gênero'] )
a5 = cramer_v( a['Tipo_de_assinatura'], a['Tipo_de_assinatura'] )
a6 = cramer_v( a['Tipo_de_assinatura'], a['Duração_do_contrato'] )

a7 = cramer_v( a['Duração_do_contrato'], a['Gênero'] )
a8 = cramer_v( a['Duração_do_contrato'], a['Tipo_de_assinatura'] )
a9 = cramer_v( a['Duração_do_contrato'], a['Duração_do_contrato'] )


# Conjunto de dados final
d = pd.DataFrame( {'Gênero': [a1, a2, a3],
                'Tipo_de_assinatura': [a4,a5,a6],
                'Duração_do_contrato': [a7,a8, a9],})

d = d.set_index( d.columns )

sns.heatmap( d, annot=True );
```

## 12.2 Ajuste (Tuning) dos parâmetros com RandomizedSearchCV e Validação Cruzada

> * Existem três estratégias para encontrar o conjunto de parâmetros que maximiza a aprendizagem do modelo: 

**1**. **Random Search**: Define os valores para cada um dos parâmetros de forma aleatória. Uma das vantagens é que ele mais rápido se comparado com as outras técnicas. Porém, dentre as desvantagens está o fato de que ele acaba selecionando o mesmo conjunto de valores que já foi testado em outra interação. Além disso, uma outra desvantagem é que ele pode nunca conseguir encontrar o melhor conjunto de valores que maximiza a aprendizagem do modelo, porque como a pesquisa é aleatória pode ser que por causa dessa aleatoriedade ele não encontre o melhor conjunto de parâmetros.

**2**. **Grid Search**: Define todas as combinações possíveis de valores que os parâmetros podem assumir. A vantagem é que ele realmente consegue encontrar o melhor conjunto de valores para cada um dos parâmetros que maximiza o modelo e, caso ele não encontre, fica próximo disso. Contudo, pelo fato de ele buscar fazer todas as combinações possíveis, demora muito tempo.

**3**. **Bayesian**: Define os valores dos parâmetros seguindo a teoria de Bayes. Ele busca encontrar o melhor conjunto de parâmetros baseado na aprendizagem anterior. Ele não faz nada aleatório como o Random Search e também não testa todas as possibilidades como o Grid Search, mas busca encontrar o melhor conjunto baseado na aprendizagem anterior. Ele é mais rápido do que o Grid Search, porém é mais lento do que o Random Search. 

* Por ser mais rápido que os demais, utilizei o Random Search. Porém, obtive êxito somente na árvore de decisão, random forest, redes neurais e regressão logística. ### 12.2.4 Implementação da Validação Cruzada

* Usamos as classes cross_val_score e  KFold  para fazer a implementação.


* Abrimos uma nova célula de código conforme abaixo e criamos uma variável resultado com cada um dos algoritmos.


* Depois, a ideia era fazer um for 30. Isso indica que iriamos rodar 30 testes. Esse valor é muito aceito na comunidade científica. Ou seja, você executa 30 testes com validação cruzada. Contudo, demorou muito tempo, então resolvi colocar somente 5.


* Cada vez que executamos a validação cruzada, temos 10 resultados, se utilizarmos o k=10. Esse é um número bem aceito. Porém, depois da demora decidir colocar k=2.


* Multiplicando 2 por 5 que é o número de vezes que vamos executar, isto é, 2 * 5 = 20, indica que teremos um total de 20 testes para cada um dos algoritmos. Antes havíamos trabalhado com somente 1 teste, agora vamos trabalhar com 20 testes por algoritmo. Depois faremos a extração da média e dessa forma teremos um resulado muito mais confiável do que o anterior.


* Criamos uma variável kfold e instânciamos o objeto dessa classe, isto é, kfold. Definimos o parâmetro n_splits=2 que é o número de divisões que você vai fazer na base de dados. O valor padrão é 10, mas decidir colocar 2 para rodar mais rápido. Isso indica que a base de dados será dividida em 2 pedaços e a cada rodada nós utilizamos 1 pedaços para treinamento e 1 pedaço para teste. O shuffle=True é para que o algoritmo sempre misture esses dados. O random_state=i indica que a cada rodada do for passaremos um valor diferente, o que significa que a cada vez que executarmos teremos resultados diferentes. Se por exemplo, colocarmos random_state=1, toda vez que rodamos sempre teremos a mesma divisão na base de dados e, consequentemente, os resultados sempre serão os mesmos e aí teríamos que sempre executar vários testes para ter resultados diferentes.


* Agora criamos cada algoritmo individualmente e passamos os parâmetros conforme havíamos feito no Tuning. Ou seja, primeiramnete fazemos o Tuning e depois criamos o algoritmo com os melhores parâmeros que foram obtidos. Depois executamos o processo de validação cruzada. Para criar esse processo de validação cruzada passamos a variável  scores e chamamos a função cross_val_score, depois passamos o algoritmo, atributos previsores, a classe e cv = kfold. Essa variável kfold serve para dividir a base de dados e a função cross_val_score recebe esses dados dívidos e vai fazer todos os testes.

```python
# Processo de Implementação da Validação Cruzada

resultados_arvore = []
resultados_random_forest = []
resultados_knn = []
resultados_logistica = []
resultados_svm = []
resultados_rede_neural = []
resultados_naive_bayes = []

for i in range(5):
  print(i)
  kfold = KFold(n_splits=2, shuffle=True, random_state=i)

  arvore = DecisionTreeClassifier(criterion='entropy', min_samples_leaf=1, min_samples_split=10, splitter='best') # criamos cada algoritmo individualmente e passamos os parâmetros conforme havíamos feito no Tuning
  scores = cross_val_score(arvore, X_churn, y_churn, cv = kfold) # Executando o processo de validação cruzada. Para criar esse processo de validação cruzada passamos a variável  scores e chamamos a função cross_val_score, depois passamos o algoritmo, atributos previsores, a classe e cv = kfold.
  #print(scores)
  #print(scores.mean())
  resultados_arvore.append(scores.mean())

  random_forest = RandomForestClassifier(criterion = 'gini', min_samples_leaf = 1, min_samples_split=10, n_estimators = 40)
  scores = cross_val_score(random_forest, X_churn, y_churn, cv = kfold)
  resultados_random_forest.append(scores.mean())

  knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p = 2)
  scores = cross_val_score(knn, X_churn, y_churn, cv = kfold)
  resultados_knn.append(scores.mean())

  logistica = LogisticRegression(C = 2.0, solver = 'saga', tol = 1e-05)
  scores = cross_val_score(logistica, X_churn, y_churn, cv = kfold)
  resultados_logistica.append(scores.mean())

  svm = SVC(kernel='rbf', random_state=1, C = 2.0)
  scores = cross_val_score(svm, X_churn, y_churn, cv = kfold)
  resultados_svm.append(scores.mean())

  rede_neural = MLPClassifier(activation = 'relu', batch_size = 56, solver = 'adam')
  scores = cross_val_score(rede_neural, X_churn, y_churn, cv = kfold)
  resultados_rede_neural.append(scores.mean())

  naive_bayes = GaussianNB()
  scores = cross_val_score(naive_bayes, X_churn, y_churn, cv = kfold)
  resultados_naive_bayes.append(scores.mean())
```

```python
# Visualizando os resultados

resultados = pd.DataFrame({'Árvore de decisão': resultados_arvore, 'Random forest': resultados_random_forest,
                           'KNN': resultados_knn, 'Regressão Logística': resultados_logistica,
                           'SVM': resultados_svm, 'Rede neural': resultados_rede_neural, 'Naive Bayes':resultados_naive_bayes})
resultados
```

## 12.3 Teste de normalidade nos resultados

* O objetivo dos testes estatísticos é saber se um algoritmo é melhor do que outro.


* Desse modo, vamos fazer testes estatísticos, e esses conceitos são muito importantes caso você queira colocar um novo modelo em produção para uma empresa. Você precisa ter a certeza que um algoritmo é melhor do que outro para você fazer a escolha certa. E também, caso for fazer artígos científicos que você queira fazer testes com algoritmos você precisa provar estatísticamente que um algoritmo é melhor do que o outro. Ou seja, vamos supor que você inventou um novo algoritmo e quer provar que ele é melhor do que os algoritmos já existentes e a maneira de fazer isso é utilizando os testes estatísticos.


* Para aplicar o teste de hipótese obrigatoriamente os dados precisam estar em uma distribuição normal.

### 12.3.1 Teste de Shapiro

* Shapiro: https://en.wikipedia.org/wiki/Shapiro%E2%80%93Wilk_test

```python
# Executando o teste shapiro passando os resultados de cada algoritmo.

shapiro(resultados_arvore), shapiro(resultados_random_forest), shapiro(resultados_knn), shapiro(resultados_logistica), shapiro(resultados_svm), shapiro(resultados_rede_neural), shapiro(resultados_naive_bayes)
```

* Sendo assim, o p-value de todos os algorítmos não é menor ou igual ao alpha. Indica que nós aceitamos a hipótese nula. Em outras palavras todos esses dados estão na distribuição normal.

### 12.3.3 Teste de hipótese com ANOVA e Tukey

* Esse teste pode ser aplicado quando temos dados que estão em uma distribuição normal. Segundo o teste de Shapiro, todos os resultados de todos os algoritmos estão numa distribuição normal.

#### >>> Aplicação do Teste Anova

* O teste Anova nesse nosso contexto vai nos ajudar a decidirmos se os resultados são diferentes, pois temos um conjunto com alguns algoritmos e precisamos primeiramente indicar se existe uma difereça estatística entre eles.

```python
# Implementando o Teste Anova

_, p = f_oneway(resultados_arvore, resultados_random_forest, resultados_knn, resultados_logistica, resultados_svm, resultados_rede_neural, resultados_naive_bayes)
p
```

```python
alpha = 0.05
if p <= alpha:
  print('Hipótese nula rejeitada. Dados são diferentes')
else:
  print('Hipótese alternativa rejeitada. Resultados são iguais')
```

* Criamos a variável p que recebe a função f_oneway e passamos os resultados de cada um dos valores.


* Essa função nos retorna dois valores. O primeiro não nos interessa, pois queremos saber somente o valor de p.


* Temos um valor em notação científica. Isso significa que rejeitamos a hipóse nula, portanto há diferença significativa entre os resultados dos algoritmos. Ou seja, quer dizer que os resultados das redes neurais é diferente do que os resultados do random forest, diferente do resultado da regressão logística e, assim, por diante.


* E chegando a conclusão de que os dados são diferentes, podemos fazer a aplicação de um outro teste que é o teste Tukey para verificar qual é o melhor algoritmo. Se a hipótese nula fosse aceita, ou seja, os resultados dos algoritmos fossem iguais não havendo diferença estatística entre ele, poderíamos chegar a conclusão que tanto faz utilizar qualquer um dos algoritmos, ou seja, tanto faz utilizar a rede neural, ou árvore de decisão, ou regressão logística e, assim, por diante. Contudo, como há diferença estatística entre os resultados dos algoritmos, precisamos verificar qual é o melhor algoritmo.

#### >>> Aplicação do Teste Tukey

* Criamos a variável resultados_algoritmos. Criamos um DataFrame e passando a taxa de acerto e concatenando todos os resultados. Concatenamos com o nome do algoritmo. Como temos 5 resultados para cada algoritmo, concatenamos 5 vezes para cada algoritmo.

```python
# Criamos uma variável que recebe MultiComparison e passamos os resultados.
# Ele faz o agrupamento da taxa de acerto pelo resultado os agoritmos

compara_algoritmos = MultiComparison(resultados_df['accuracy'], resultados_df['algoritmo'])
```

## 12.5 Combinação e Rejeição de classificadores

### 12.5.1 Combinação de classificadores

```python
# Já temos essa codificação pré-definida que é exatamente o mesmo que havíamos implementado anteriormente.
# Buscamos um novo registro na base de dados na posição 0, fazemos o reshape(1, -1) para colocar no formato de matriz.

novo_registro = X_churn[0]
novo_registro = novo_registro.reshape(1, -1)
novo_registro, novo_registro.shape
```

```python
# Combinando os classificadores

cancela_servico = 0
não_cancela_servico = 0

if resposta_arvore_decisao[0] == 1:
  não_cancela_servico += 1
else:
  cancela_servico += 1

if resposta_random_forest[0] == 1:
  não_cancela_servico += 1
else:
  cancela_servico += 1

if resposta_rede_neural[0] == 1:
  não_cancela_servico += 1
else:
  cancela_servico += 1

if cancela_servico > não_cancela_servico:
  print('Cliente cancelará o serviço')
elif cancela_servico == não_cancela_servico:
  print('Empate')
else:
  print('Cliente não cancelará o serviço')
```

### 12.5.2 Rejeição de classificadores

```python
# Já temos essa codificação pré-definida que é exatamente o mesmo que havíamos implementado anteriormente.
# Buscamos um novo registro na base de dados na posição 0, fazemos o reshape(1, -1) para colocar no formato de matriz.

novo_registro = X_churn[0]
novo_registro = novo_registro.reshape(1, -1)
novo_registro, novo_registro.shape
```

```python
# Obtendo as respostas (previsão) para cada um dos algorítmos

resposta_arvore_decisao = classificador_arvore_decisao.predict(novo_registro)
resposta_random_forest = classificador_random_forest.predict(novo_registro)
resposta_rede_neural = classificador_rede_neural.predict(novo_registro)

resposta_arvore_decisao [0], resposta_random_forest[0],resposta_rede_neural[0]
```

```python
# Probabilidade da Árvore de Decisão

probabilidade_arvore_decisao = classificador_arvore_decisao.predict_proba(novo_registro)
probabilidade_arvore_decisao
```

```python
# Obtendo o maior valor de probabildade da Árvore de Decisão que foi previsto

confianca_arvore_decisao = probabilidade_arvore_decisao.max()
confianca_arvore_decisao
```

```python
# Obtendo o maior valor de probabildade do Random Forest que foi previsto

probabilidade_random_forest = classificador_random_forest.predict_proba(novo_registro)
confianca_random_forest = probabilidade_random_forest.max()
confianca_random_forest
```

```python
# Obtendo o maior valor de probabildade da Rede Neural que foi previsto

probabilidade_rede_neural = classificador_rede_neural.predict_proba(novo_registro)
confianca_rede_neural = probabilidade_rede_neural.max()
confianca_rede_neural
```

```python
# Rejeitando os classificadores

cancela_servico = 0
não_cancela_servico = 0
confianca_minima = 0.999999
algoritmos = 0

if confianca_arvore_decisao >= confianca_minima:
  algoritmos += 1
  if resposta_arvore_decisao[0] == 1:
    não_cancela_servico += 1
  else:
    cancela_servico += 1

if confianca_random_forest >= confianca_minima:
  algoritmos += 1
  if resposta_random_forest[0] == 1:
    não_cancela_servico += 1
  else:
    cancela_servico += 1

if confianca_rede_neural >= confianca_minima:
  algoritmos += 1
  if resposta_rede_neural[0] == 1:
    não_cancela_servico += 1
  else:
    cancela_servico += 1

if cancela_servico > não_cancela_servico:
  print('Cliente cancelará o serviço, baseado em {} algoritmos'.format(algoritmos))
elif cancela_servico == não_cancela_servico:
  print('Empate, baseado em {} algoritmos'.format(algoritmos))
else:
  print('Cliente não cancelará o serviço, baseado em {} algoritmos'.format(algoritmos))
```

**2**. **Uma outra maneira é através do próprio modelo de marchine learning verificar o incremento que o modelo trás para o negócio, seja incremento de receita, redução de custo, redução de tempo, perda de clientes, dentre outros.**:
